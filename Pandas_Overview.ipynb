```python
# Exploring data in pandas

import pandas as pd
import numpy as np

data = np.random.uniform(10, 50, (7,8))
df_large = pd.DataFrame(data, columns=[f'col_{i}' for i in range(8)])
print(df_large)
print(df_large.head())
print(df_large.tail())
print(df_large.info())
print(df_large.describe())
print(df_large.dtypes)
print(df_large.shape)
print(df_large.columns)
print(df_large['col_1'].unique())
print(df_large['col_3'].nunique())
print(df_large['col_5'].value_counts())
```

```python
# Indexing and Selection in Pandas
# loc
data = np.random.rand(3,4)
df = pd.DataFrame(data, columns=['A','B','C','D'], index=[1,2,3])
print(df)
print(df.loc[1])
print(df.loc[[1,2]])
print(df.loc[1,'B'])
print(df.loc[[1,2],['A']])
print(df.loc[:,['C']])
print(df.loc[1:3,['A','B']])
```

```python
# iloc
print(df)
print(df.iloc[1])
print(df.iloc[[0,2]])
print(df.iloc[1,3])
print(df.iloc[0:2,0:2])
print(df.iloc[1:3,:])
```

```python
import numpy as np
import pandas as pd

# Creating a DataFrame for conditional selection
data = np.random.uniform(3,50, (3,4))
df = pd.DataFrame(data, columns=['A','B','C','D'], index=[1,2,3])

print(df)
print(df[df > 25])
print(df[df['A'] > 25])
print(df[(df['A'] > 25) & (df['B'] < 10)])
print(df[(df['A'] > 25) | (df['B'] < 10)])
print(df.query('A > 25 and B > 20'))
print(df.query('not (A > 25 and B > 20)'))
print(df[~(df['A'] < 40)])
print(df[df['C'].isin([25,30])])
```
