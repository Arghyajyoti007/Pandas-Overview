### Data Cleaning Steps Performed:

1.  **Standardized Column Names**: All column names were stripped of leading/trailing spaces, converted to uppercase, and spaces were replaced with underscores.
2.  **Renamed Column**: The `DATE_OF_JOINING` column was renamed to `JOINING_DATE`.
3.  **Filled Missing First/Last Names**: Missing values in the `FIRST_NAME` column were filled with "First Name not available", and missing values in the `LAST_NAME` column were filled with "Last Name not available".
4.  **Removed Rows with Missing Employee IDs**: Rows where `EMPLOYEE_ID` was null were dropped from the DataFrame.
5.  **Dropped 'JOINING_DATE' Column**: The entire `JOINING_DATE` column was removed.
6.  **Cleaned and Filled 'SALARY' Column**:
    *   Non-numeric values in the `SALARY` column were converted to `NaN`.
    *   Missing `SALARY` values (`NaN`) were filled with the mean of the numeric salaries.
    *   The `SALARY` column was then converted to an integer data type.
7.  **Filled Missing Email Addresses**: Missing values in the `EMAIL` column were filled with "Email address not found".

```python
import pandas as pd
# read data
df = pd.read_excel("/content/uncleaned_employee_data.xlsx")

# check data
print(df.info())
print(df.head(10))
print(df.isnull().sum())

# maintain uniformity
df.columns = df.columns.str.strip().str.upper().str.replace(" ","_")
print(df.columns)

# Rectify Column Names
df.rename(columns={"DATE_OF_JOINING": "JOINING_DATE"}, inplace=True)
print(df.columns)

# Recify Nulls by filling and removing
df['FIRST_NAME'] = df['FIRST_NAME'].fillna("First Name not available")
df['LAST_NAME'] = df['LAST_NAME'].fillna("Last Name not available")

df.dropna(subset=['EMPLOYEE_ID'], inplace=True)
print(df.head(10))
print(df.isnull().sum())

df.drop("JOINING_DATE", axis=1, inplace = True)
print(df.head(10))
print(df.isnull().sum())

# Convert non-numeric values to NaN and then to numeric type
df['SALARY'] = pd.to_numeric(df['SALARY'], errors='coerce')

# Calculate mean after converting to numeric
mean_salary = df['SALARY'].mean()
df['SALARY'] = df['SALARY'].fillna(mean_salary)

# Now convert to integer type
df['SALARY'] = df['SALARY'].astype(int)

print(df.head(10))
print(df.isnull().sum())

df['EMAIL'] = df['EMAIL'].fillna("Email address not found")
print(df.head(10))
print(df.isnull().sum())

# remove duplicates
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())

# export data
df.to_excel("cleaned_employee_data.xlsx", index=False)
```
